# Выделение ROI с помощью ключевых точек рук

ROI (Region of Interest) - в компьютерном зрении и оптическом распознавании символов ROI определяет границы рассматриваемого объекта. То есть буквально мы извлекаем из картинки только то, что нас интересует. Такая предобработка помогает существенно снизить вычислительные затраты, так как уменьшается размер изображения, а также улучшить точность прогнозов.

```python
DATA_PATH = "/content/drive/MyDrive/DATA_TOTAL_01_06"
#DATA_PATH = "путь до пользовательских картинок"
RANDOM_STATE = 128


# Создадим датафрейм, где будем сохранять путь до картинки и ее лейбл
dactyl = ["А", "Б", "В", "Г", "Е",
          "Ж", "И", "К", "Л", "М",
          "Н", "О", "П", "Р", "С",
          "Т", "У", "Ф", "Х", "Ч",
          "Ш", "Ы", "Э", "Ю", "Я"]

list_path = []
list_labels = []
for label in dactyl:
    label_path = os.path.join(DATA_PATH, label)
    for files in os.listdir(label_path):
      list_path.append(files)
      list_labels.append(label)

metadata = pd.DataFrame({
    "image_path": list_path,
    "label": list_labels
})

metadata.head()
```

# Сначала попробуем обрезать изображения по крайним координатам

```python
detector = HandDetector(maxHands = 1, staticMode=True, detectionCon=0.1)

offset = 20
imgSize = 300
counter = 0

def get_image(img_path):
  """
    Process an image to extract and resize the hand region.

    Input:
        img_path (str): Path to the input image file.

    Output:
        np.array: Processed image with the hand region resized and centered in a 300x300 white background.
                  Returns None if the image does not contain a hand or if the image is not found.
  """
  img = cv2.imread(img_path)
  if img is None:
    print(f'image {img_path} do not have landmarks')
    return img

  hands, img = detector.findHands(img, draw=False)
  if hands:
    hand = hands[0]
    x,y,w,h = hand['bbox']
    imgWhite = np.ones((imgSize, imgSize,3),np.uint8)*255
    imgCrop = img[y-offset:y+h+offset,x-offset:x+w+offset]
    imgCropShape = imgCrop.shape

    aspectRatio = h/w
    if aspectRatio >1: #for width
      k = imgSize/h #contastant k
      wCal= math.ceil(k*w) #caluclatated width
      imgResize = cv2.resize(imgCrop, (wCal,imgSize))
      imgResizeShape = imgResize.shape
      wGap = math.ceil((imgSize- wCal)/2)#width gap
      imgWhite[:,wGap:wCal+wGap] = imgResize
      return imgCrop

    else: #for height
      k = imgSize/w
      hCal= math.ceil(k*w)
      imgResize = cv2.resize(imgCrop, (imgSize, hCal))
      imgResizeShape = imgResize.shape
      hGap = math.ceil((imgSize- hCal)/2)

      imgWhite[hGap:hCal+hGap,:] = imgResize
      return imgCrop

  else:
    print(f'image {img_path} does not have hands')


print("Пример использования")
img_path = "/content/drive/MyDrive/DATA_TOTAL_01_06/О/881416533_AgACAgIAAxkBAAIUCGYjZFDesnhgNQvecPBptE9JvGpkAAKn2zEb1CQZSaZ5Mo7Ycl0JAQADAgADeQADNAQ.jpg"
processed_img = get_image(img_path)
if processed_img is not None:
    plt.imshow(processed_img)
```

```python
train_ratio = 0.90
test_ratio = 0.10
#validation_ratio = 0.10

x_train, x_test, y_train, y_test = train_test_split(metadata.image_path,
                                                    metadata.label,
                                                    test_size= 1 - train_ratio,
                                                    random_state=RANDOM_STATE,
                                                    shuffle=True,
                                                    stratify=metadata.label)


x_train.reset_index(inplace=True, drop=True)
y_train.reset_index(inplace=True, drop=True)

x_test.reset_index(inplace=True, drop=True)
y_test.reset_index(inplace=True, drop=True)
```

```python
dir_name_train = "папка_для_новых_фото_TRAIN"
num_droped_images = 0

for i in range(len(x_train)):
  empty_array = get_image(os.path.join(DATA_PATH, y_train[i], x_train[i]))
  if empty_array is None:
    num_droped_images += 1
    continue
  else:
    cv2.imwrite(os.path.join(DATA_PATH, dir_name_train, y_train[i], x_train[i]), empty_array)
    if i%100 == 0:
      print(f"Обработано {i} изображений")

print(f"Препроцессинг для выборки x_train ЗАКОНЧЕН")
print(f"Кол-во выброшенных данных: {num_droped_images}")
```

```python
dir_name_test = "папка_для_новых_фото_TEST"
os.mkdir(os.path.join(DATA_PATH, dir_name_test))

for symbol in set(y_test):
  os.mkdir(os.path.join(DATA_PATH, dir_name_test, symbol))

num_droped_images = 0
for i in range(len(x_test)):
  empty_array = get_image(os.path.join(DATA_PATH, y_test[i], x_test[i]))
  if empty_array is None:
    num_droped_images += 1
    continue
  else:
    cv2.imwrite(os.path.join(DATA_PATH, dir_name_test, y_test[i], x_test[i]), empty_array)
    if i%10 == 0:
      print(f"Обработано {i} изображений")

print(f"Препроцессинг для выборки x_test ЗАКОНЧЕН")
print(f"Кол-во выброшенных данных: {num_droped_images}")
```
